```python
from matplotlib import pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
x2=[1,2,3,4,5]
y2=[5,4,3,2,1]
plt.plot(x,y,"r")
plt.plot(x2,y2,"bp")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
x2=[1,2,3,4,5]
y2=[5,4,3,2,1]
plt.plot(x,y,"r")
plt.plot(x2,y2,"b")
plt.show()
```

```python
import numpy as np
x = np.arange(1,10,0.1)
y=x*2+1
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=x*2+1
x2=np.array([5,4,3,2,1])
y2=x2*2+1
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p")
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p",linewidth=3)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p",linewidth=0.2)
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p",linewidth=0.2,label="x^4")
plt.legend()
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p",linewidth=0.2,label="x^4")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p",linewidth=0.2,label="x^4")
plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.legend()
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p",linewidth=0.2,label="x^4")
plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.legend()
plt.axis([0,20,0,1000])
plt.grid()
plt.show()
```

```python
x=np.arange(0,10,0.25)
y=x**4
plt.plot(x,y, color="red" , marker="p",linewidth=0.2,label="x^4")
plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.legend()
plt.axis([0,20,0,1000])
plt.title("chart")
plt.text(15,800,"Keerthi",fontsize=15)
plt.grid()
plt.show()
```
